```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/content/drive/MyDrive/lfwcrop_grey/rosee.jpg')

# Perform arithmetic operations on the image
result_addition1 = cv2.add(image, 50)
result_addition2 = cv2.add(image, 100)
result_subtraction1 = cv2.subtract(image, 50)
result_subtraction2 = cv2.subtract(image, 100)
result_multiplication1 = cv2.multiply(image, 0.5)
result_multiplication2 = cv2.multiply(image, 1.5)
result_inversion = cv2.bitwise_not(image)

# Display the original and processed images using matplotlib
plt.figure(figsize=(5, 10))

# Addition Result 1
plt.subplot(4, 2, 1)
plt.imshow(cv2.cvtColor(result_addition1, cv2.COLOR_BGR2RGB))
plt.title('Addition Result 1')
plt.axis('off')

# Addition Result 2
plt.subplot(4, 2, 2)
plt.imshow(cv2.cvtColor(result_addition2, cv2.COLOR_BGR2RGB))
plt.title('Addition Result 2')
plt.axis('off')

# Subtraction Result 1
plt.subplot(4, 2, 3)
plt.imshow(cv2.cvtColor(result_subtraction1, cv2.COLOR_BGR2RGB))
plt.title('Subtraction Result 1')
plt.axis('off')

# Subtraction Result 2
plt.subplot(4, 2, 4)
plt.imshow(cv2.cvtColor(result_subtraction2, cv2.COLOR_BGR2RGB))
plt.title('Subtraction Result 2')
plt.axis('off')

# Multiplication Result 1
plt.subplot(4, 2, 5)
plt.imshow(cv2.cvtColor(result_multiplication1, cv2.COLOR_BGR2RGB))
plt.title('Multiplication Result 1')
plt.axis('off')

# Multiplication Result 2
plt.subplot(4, 2, 6)
plt.imshow(cv2.cvtColor(result_multiplication2, cv2.COLOR_BGR2RGB))
plt.title('Multiplication Result 2')
plt.axis('off')

# Original Image
plt.subplot(4, 2, 7)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Inversion Result
plt.subplot(4, 2, 8)
plt.imshow(cv2.cvtColor(result_inversion, cv2.COLOR_BGR2RGB))
plt.title('Inversion Result')
plt.axis('off')

plt.show()
```
